```python
row = []
i = 0
for u in users['members']:
    row.append(u['id'])
    row.append(u['profile']['display_name'])
    if 'email' in u['profile']:
        print(u['profile']['email'])
        row.append(u['profile']['email'])
    else:
        print('n/a')
        row.append('n/a')
    if 'phone' in u['profile']:
        print(u['profile']['phone'])
        row.append(u['profile']['phone'])
    else:
        print('n/a')
        row.append('n/a')
    df.loc[i] = row
    i+=1
    print(row)
    row = []
```
